```python
import pandas as pd
import seaborn as sns
import numpy as np
```

```python
tracks = pd.read_csv('data/joined_mod5_project.csv')
tracks.info()
```

```python
#marking tracks with genre "pop"
tracks['pop'] = [1 if ("pop" in x) else 0 for x in tracks.artist_genres] 
tracks['pop'].value_counts()
```

```python
#Creating a new dataframe
tracks_pop = tracks.loc[tracks['pop']==1].copy()
tracks_pop.drop(['Unnamed: 0','pop'],axis=1,inplace=True)
```

```python
tracks_pop['artist_genres'] = [x.split(',') for x in tracks_pop['artist_genres']]
```

```python
for row in tracks_pop['artist_genres']:
    for subgenre in row:
        row[:] = [value for value in row if 'pop' in value]
```

```python
from operator import itemgetter

subgenre_dict = dict()
for row in tracks_pop['artist_genres']:
    for subgenre in row:
        if subgenre in subgenre_dict:
            subgenre_dict[subgenre] += 1
        else:
            subgenre_dict[subgenre] = 1

sorted(subgenre_dict.items(), key=itemgetter(1), reverse=True)
```
