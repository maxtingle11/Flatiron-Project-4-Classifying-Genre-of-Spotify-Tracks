```python
import pandas as pd
import numpy as np

pd.set_option('display.max_rows', 50)
pd.set_option('display.max_columns', 400)
pd.set_option('display.width', 1000)

songs = pd.read_csv('data/spotify_tracks.csv')
songs = songs.drop(["Unnamed: 0"], axis=1)
print(songs.info())
songs.head()
```

```python
# Create Pop Songs DataFrame

# Create pop column of NaNs
songs['pop'] = np.nan

# Create boolean True if 'pop' is contained in artist_genres
songs['pop'] = songs.artist_genres.apply(lambda x: 'pop' in x)

# Create pop_songs dataframe
pop_songs = songs.loc[songs['pop'] == True]

# Re-set index
pop_songs.reset_index(inplace=True)

# Drop unnecessary columns
pop_songs = pop_songs.drop(['index', 'pop'], axis=1)

# Transform strings in 'artist_genres' column to lists
pop_songs['artist_genres'] = pop_songs.artist_genres.apply(lambda x: x.split(','))

# Copy dataframe to use later on
pop_songs2 = pop_songs.copy()

print(pop_songs.info())
pop_songs.head()
```

```python
# Create list of unique genres

# Create array of lists from artist_genre column
genres_array = pop_songs.artist_genres.values

# Create a set of unique genres
unique_genres = set(x for l in genres_array for x in l)

# Convert set to list
unique_genres = list(unique_genres)

# Sort list of unique genres
unique_genres.sort()

# Count of unique subgenres
len(unique_genres)
```

```python
# Create list of pop subgenres with for-loop
pop_subgenres = []
for genre in unique_genres:
    if "pop" in genre:
        pop_subgenres.append(genre)

len(pop_subgenres)
```

```python
# Creating new columns for each pop subgenre and assign NaN values
for genre in pop_subgenres:
    pop_songs[genre] = np.nan

# Assigning True Boolean value to pop subgenre columns 
for genre in pop_subgenres:
    pop_songs[genre] = pop_songs.artist_genres.apply(lambda x: genre in x)
```

```python
# Create list of dictionaries with counts for each pop_subgenre
pop_subgenre_counts = []
for genre in pop_subgenres:
    info = {'pop_subgenre': genre, 'count': pop_songs[genre].sum()}
    pop_subgenre_counts.append(info)
```

```python
# Create dataframe of pop_subgenres, and sort by count descending 
pop_subgenre_counts_df = pd.DataFrame(pop_subgenre_counts)
pop_subgenre_counts_df = pop_subgenre_counts_df.sort_values(by=['count'], ascending=False)
pop_subgenre_counts_df.head(20)
```

```python
# List of top 7
top_pop_subgenres = ['dancepop', 
                     'poprap', 
                     'folkpop', 
                     'electropop', 
                     'poprock', 
                     'indiepop', 
                     'popedm']
```

```python
# Creating new columns for each pop subgenre and assign NaN values
for genre in top_pop_subgenres:
    pop_songs2[genre] = np.nan

# Assigning True Boolean value to pop subgenre columns 
for genre in top_pop_subgenres:
    pop_songs2[genre] = pop_songs2.artist_genres.apply(lambda x: genre in x)
```

```python
# Create new count column and fill with NaNs
pop_songs2['top_genre_count'] = np.nan
```

```python
#Create column that counts number of subgenres per row that are True
pop_songs2['top_genre_count'] = pop_songs2.iloc[:,19:26].apply(lambda s: (s > 0).sum(), axis=1)
```

```python
# See how many songs fit into 2+ subgenres
pop_songs2['top_genre_count'].value_counts()
```

```python
# Take out songs with more than one subgenre
pop_songs2_filtered = pop_songs2[(pop_songs2['top_genre_count']<=1)]
```

```python
# See how many songs fit into 2+ subgenres
pop_songs2_filtered.top_genre_count.value_counts()
```

```python
pop_songs2_filtered['dancepop'] = [1 if x == True else 0 for x in pop_songs2_filtered['dancepop']]
pop_songs2_filtered['poprap'] = [2 if x == True else 0 for x in pop_songs2_filtered['poprap']]
pop_songs2_filtered['folkpop'] = [3 if x == True else 0 for x in pop_songs2_filtered['folkpop']]
pop_songs2_filtered['electropop'] = [4 if x == True else 0 for x in pop_songs2_filtered['electropop']]
pop_songs2_filtered['poprock'] = [5 if x == True else 0 for x in pop_songs2_filtered['poprock']]
pop_songs2_filtered['indiepop'] = [6 if x == True else 0 for x in pop_songs2_filtered['indiepop']]
pop_songs2_filtered['popedm'] = [7 if x == True else 0 for x in pop_songs2_filtered['popedm']]
```
